```python
min_length = min(len(data['Country']), len(data['Continent']), len(data['Average Salary']))
data = {key: value[:min_length] for key, value in data.items()}
```

```python
df = pd.DataFrame(data)
```

```python
df['Average Salary'] = df['Average Salary'].replace('[\$,]', '', regex=True).astype(float)
```

```python
pivot_df = df.pivot(index="Country", columns="Continent", values="Average Salary")
```

```python
plt.figure(figsize=(14, 18))
sns.heatmap(pivot_df, cmap="YlGnBu", annot=False)
plt.title('Mapa de Calor de Salarios Promedio por País')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.boxplot(x='Continent', y='Average Salary', data=df)
plt.title('Distribución de Salarios Promedio por Continente')
plt.xticks(rotation=45)
plt.show()
```
